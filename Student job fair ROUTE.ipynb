# ***Feature Engineering***

```python
data['total_score'] = data['experience_years'] + data['course_grades'] + data['projects_completed'] + data['extracurriculars']
```

```python
data['mean_course_grades'] = data['course_grades'].mean()
```

```python
data
```

```python
x = data.drop(['job_offer'], axis = 1)
y = data['job_offer']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

# ***Scaling***

```python


from sklearn.preprocessing import MinMaxScaler

mn = MinMaxScaler(feature_range=(-1, 1))
data['total_score'] = mn.fit_transform(data[['total_score']])
data['mean_course_grades'] = mn.fit_transform(data[['mean_course_grades']])
data['projects_completed'] = mn.fit_transform(data[['projects_completed']])
data['experience_years'] = mn.fit_transform(data[['experience_years']])
data['course_grades'] = mn.fit_transform(data[['course_grades']])
```

```python
x_train_sc = mn.fit_transform(x_train)
x_test_sc = mn.transform(x_test)
```

# ***Feature Selection***

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
fs = SelectFromModel(RandomForestClassifier())
fs.fit(x_train_sc, y_train)
```

# ***Deep Learning modeling***

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
model = Sequential([
    layers.Dense(64, activation= 'relu', input_shape= (4,)),
    layers.Dense(32, activation= 'relu'),
    layers.Dense(16, activation= 'relu'),
    layers.Dense(8, activation= 'relu'),
    layers.Dense(1, activation= 'sigmoid')
])
```

```python
model.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
```

```python
x_train_sc_modified = x_train_sc[:, :4]
x_train_sc_modified = x_train_sc_modified.astype(np.float32)  # Convert to float32 if needed
y_train = y_train.astype(np.float32)  # Convert to float32 if needed
```

```python
model.fit(x_train_sc_modified, y_train, epochs = 150)
```

# ***Model Evaluation***

```python
#model evaluation in Deep Learing

x_test_sc = x_test_sc[:, :4]

model.evaluate(x_test_sc, y_test)
```

```python
#model evaluation in Machine Learning

from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test_sc)
y_pred = y_pred.round()
accuracy_score(y_test, y_pred)
```
